```python
import matplotlib.pyplot as plt
import numpy as np 
import librosa
import librosa.display
import IPython.display as ipd
```

```python
n = 64
te = 1/8000
t = np.linspace(0, n*te, n,endpoint=False)
s = 0.75*np.cos(2*np.pi*250*t)
plt.figure(figsize=(14, 5))
plt.plot(t,s)
plt.show()
```

```python
n = 64
te = 1/8000
t = np.linspace(0, n*te, n,endpoint=False)
s = 0.75*np.cos(2*np.pi*280*t)
plt.figure(figsize=(14, 5))
plt.plot(t,s)
plt.show()
```

```python
n = 16000
fe = 8000
te = 1/8000
t = np.linspace(0, n*te, n,endpoint=False)
s = 0.75*np.cos(2*np.pi*280*t) + 0.5*np.cos(2*np.pi*500*t)
tf = np.fft.rfftfreq(n, te)
plt.figure(figsize=(30, 5))
plt.subplot(121)
plt.plot(t, s)
plt.title("chronogramme")
plt.subplot(122)
plt.plot(tf,np.abs(np.fft.rfft(s)))
plt.title("spectre")
plt.show()
```

```python
S = np.abs(librosa.stft(s))
Sdb = librosa.amplitude_to_db(abs(S))
librosa.display.specshow(Sdb, sr=fe, x_axis='s', y_axis='log')
```

```python
s,fe=librosa.load('../Wilhelm Scream.wav')
n = len(s)
t = np.linspace(0, n/fe, n,endpoint=False)
plt.xlabel('seconds')
plt.title('chronogramme du son Wilhelm Scream.wav')
plt.plot(t,s)
```

```python
S=np.abs(np.fft.rfft(s))
tf=np.fft.rfftfreq(len(s),1/fe)
plt.xlabel('frequency (Hz)')
plt.title('spectre du son Wilhelm Scream.wav')
plt.plot(tf,S)
plt.axis([1000, 2000, 0, 350])
```
